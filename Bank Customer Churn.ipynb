# Bank Customer Churn
An example of exploratory data analysis using several **Python** libraries, including **pandas, matplotlib and seaborn**.
### Dataset
The dataset used is _Bank Customer Churn_ obtained from [Kaggle](https://www.kaggle.com/datasets/radheshyamkollipara/bank-customer-churnhttps://www.mavenanalytics.io), and it contains of a single data table (saved in csv format).

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# load the dataset into a Pandas DataFrame
df = pd.read_csv('/kaggle/input/bank-customer-churn/Customer-Churn-Records.csv')
```

```python
plt.figure(figsize=(20, 5))
sns.heatmap(df.corr(numeric_only = True), annot = True)
```

```python
df['AgeCategory'] = pd.cut(df['Age'], bins=[0, 19, 33, 46, 61, 76, np.inf],labels=['<18', '18-32', '33-45', '46-60', '60-75', '>75'])
```

```python
plt.figure(figsize=(25, 30))

plt.subplot(5, 2, 1)
sns.countplot(data=df, x='Geography', hue='Exited', palette='colorblind')
plt.title('Churn by Country')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 2)
sns.countplot(data=df, x='Satisfaction Score', hue='Exited', palette='colorblind')
plt.title('Churn by Satisfaction Score')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 3)
sns.countplot(data=df, x='Gender', hue='Exited', palette='colorblind')
plt.title('Churn by Gender')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 4)
sns.countplot(data=df, x='Tenure', hue='Exited', palette='colorblind')
plt.title('Churn by Tenure')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 5)
sns.countplot(data=df, x='NumOfProducts', hue='Exited', palette='colorblind')
plt.title('Churn by Number of Products')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 6)
sns.countplot(data=df, x='IsActiveMember', hue='Exited', palette='colorblind')
plt.title('Churn by Active Status')
plt.legend(labels = ['Retained', 'Churned'])

plt.subplot(5, 2, 7)
sns.countplot(data=df, x='AgeCategory', hue='Exited', palette='colorblind')
plt.title('Churn by Age')
plt.legend(labels = ['Retained', 'Churned'])
```
